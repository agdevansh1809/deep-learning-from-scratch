```python
import numpy as np
```

```python
def conv2d(image, kernel, stride=1, padding=0):
    """
    image: 2D numpy array (H, W)
    kernel: 2D numpy array (kH, kW)
    stride: int
    padding: int
    returns: 2D numpy array (output feature map)
    """
    # Step 1: apply padding to image if padding > 0

    # Step 2: compute output dimensions (account for stride this time, not just kernel size)

    # Step 3: initialize output array of zeros

    # Step 4: nested loop sliding the kernel, using stride as the step size

    if padding > 0:
      pad_amount = padding
      image = np.pad(image,
       ((pad_amount,pad_amount),(pad_amount,pad_amount)),
       mode = "constant",)
    h1 = image.shape[0]
    w1 = image.shape[1]

    kh = kernel.shape[0]
    kw = kernel.shape[1]

    w2 = (w1 -kw )/stride +1
    h2 = (h1 -kh)/stride +1

    w2 = int(np.round(w2))
    h2 = int(np.round(h2))


    output = np.zeros((h2,w2))

    for i in range(h2):          # slide vertically
        for j in range(w2):      # slide horizontally
            region = image[i*stride : i*stride+kh, j*stride : j*stride+kw]   # grab the patch under the kernel
            output[i, j] = np.sum(region * kernel)  # elementwise multiply, then sum

    return output

```

```python
# Test 1: matches your Day 1 hand calculation
image = np.arange(49).reshape(7,7).astype(float)
kernel = np.ones((3,3))

out = conv2d(image, kernel, stride=1, padding=0)
print(out.shape)   # should be (5,5) — you calculated this by hand earlier
print(out[0,0])    # should equal sum of image[0:3, 0:3]
```
